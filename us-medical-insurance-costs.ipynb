# U.S. Medical Insurance Costs

First import the necessary libreries. We will be working with pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import the data into Python and inspect the contents.

```python
df = pd.read_csv(r"C:/Users/Jelvy's Personal/Desktop/ML Course/Codecademy_Projects/python-portfolio-project-starter-files/insurance.csv")
print(df.head())
print(df.describe())
```

Save your dataset via Python variables to storing them

```python
age = df['age']
sex = df['sex']
bmi = df['bmi']
children = df['children']  
smoker = df['smoker']
region = df['region']      
charges = df['charges']
```

Let's print out some characteristics of our data set.

```python
print(f" The Average age of the patients is: " + str(round(age.mean(),2)) + " years old.")
```

Histogram of the distribution of ages.

```python
plt.hist(age, bins=20, color='blue', edgecolor='black')
plt.title('Age Distribution of Patients')  
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

We can conclude that people under 30 years old are overrepresented in our data.

```python
sex_counts = sex.value_counts()
male = sex_counts['male']
female = sex_counts['female']

print(f"There are {male} males and {female} females in the dataset.")
```

```python
label_counts = sex_counts.index
sizes = sex_counts.values
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=label_counts, autopct='%1.1f%%', startangle=140)
plt.title('Sex Distribution of Patients')
plt.axis('equal')
plt.show()
```

The sex representation is almost even, so this will not introduce a bias in our data.

```python
print(region.value_counts())
```

```python
region_counts = region.value_counts() 
label = region_counts.index
sizes = region_counts.values
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=label, autopct='%1.1f%%', startangle=140)
plt.title('Region Distribution of Patients')
plt.axis('equal')
plt.show()
```

Like the sex, the regions are also almost evenly represented.

```python
print(f"The average BMI of the patients is: " + str(round(bmi.mean(),2)) + " kg/m^2.")
```

Let's plot the BMI distribution.

```python
plt.hist(bmi, bins=20, color='blue', edgecolor='black')
plt.title('BMI Distribution of Patients')  
plt.xlabel('BMI')
plt.ylabel('Number of Patients')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

We get an almost normal distribution. The peak is between a BMI of 28 and 30, which indicates that the majority of our sample has a BMI of 25 or more and is overweight

Let's calculate the number of child distribution.

```python
print(children.value_counts())
```

In our sample, there is a greater number of people with one child than with two or more children. This could create a bias in the analysis, as it might suggest that people with fewer children have a higher insurance cost.

We calculate the average cost of Medical Insurance.

```python
charges_mean = charges.mean()
print(f"The average insurance cost for patients is: " + str(round(charges_mean,2)) + " dollars.")
```

Graph to determine the relationship between age and Insurance cost.

```python
plt.scatter(age, charges, alpha=0.5)
plt.axhline(charges_mean, color='red', linestyle='dashed', linewidth=2, label=f'Promedio: ${charges_mean:.2f}')
plt.title('Insurance Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

There is an upward trend in the cost of insurance as age increases. This concludes that age is a determining factor in the cost of insurance.

Graph to determine the relationship between sex and Insurance cost.

```python
plt.bar(sex, charges, alpha=0.5,width=0.5)
plt.ylim(0, 100000)
plt.title('Insurance Charges vs Sex')
plt.xlabel('Sex')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

It's clear that there isn't a significant difference between the cost of medical insurance for men and women. This leads us to conclude that sex is not as determining a factor for the cost of medical insurance.

Graph to determine the relationship between BMI and Insurance cost.

```python
plt.bar(bmi, charges, alpha=0.5)
plt.title('Insurance Charges vs BMI')
plt.xlabel('BMI')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

It can be clearly seen that the highest Insurance cost is for people with a BMI between 30 and 40. This clearly indicates a relationship between BMI and the cost of Insurance.

# Conclusions

Based on our data analysis, we can conclude that BMI, age, and the number of children are the determining factors for the cost of insurance. However, the biases found in the sample should be taken into account, as there is an overrepresentation of people with a high BMI (overweight) and people with one or no children.
